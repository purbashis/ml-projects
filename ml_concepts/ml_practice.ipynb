```python
import numpy as np

X = titanic_df["Age"].fillna(titanic_df["Age"].median()).values
y = titanic_df["Survived"].values

w = 0.0
b = 0.0
lr = 0.001

for _ in range(1000):
    y_hat = 1 / (1 + np.exp(-(w*X + b)))
    dw = np.mean((y_hat - y) * X)
    db = np.mean(y_hat - y)
    w -= lr * dw
    b -= lr * db
    


```

```python
loss = -np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))

loss
```
