```python
#CONDITIONAL PROBABILITY
#Question

#Probability of survival given passenger is female?

titanic_df[titanic_df["Sex"] =='female']['Survived'].mean()
```

```python
#BAYES THEOREM

P_female = (titanic_df['Sex']=="female").mean()
P_survived = (titanic_df['Survived'].mean())
P_survived_given_female = titanic_df[titanic_df["Sex"] == "female"]["Survived"].mean()

P_female_given_survived  = (P_survived_given_female *P_female) / P_survived

P_female_given_survived 
```

```python
import math

def nCr(n, r):
    return math.factorial(n) // (math.factorial(r) * math.factorial(n-r))

nCr(5,3)
```
