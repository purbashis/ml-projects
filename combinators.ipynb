Problem 1: Data Augmentation - Image FlipsProblem:
You have 1 image and want to create variations using:

Original
Horizontal flip
Vertical flip
Rotate 90°
Rotate 180°
Rotate 270°
How many unique images can you create if you apply exactly 2 transformations? 


```python

from itertools import permutations

transformations = [
    'Original',
    'Horizontal flip',
    'Vertical flip',
    'Rotate 90°',
    'Rotate 180°',
    'Rotate 270°'
]

# Generate all permutations of 2 transformations
#P(6,2) = 6!/(6-2)! = 6!/4! = 6 × 5 = 30 combinations
transformation_combos = list(permutations(transformations, 2))

print(f"Total unique image variations with 2 transformations: {len(transformation_combos)}\n")

for i, combo in enumerate(transformation_combos, 1):
    print(f"{i}. {combo}")
    
    
```

Problem 3: Feature Interaction Terms
Problem:
You have 8 features: [X1, X2, X3, X4, X5, X6, X7, X8]
You want to create all pairwise interaction features (X1×X2, X1×X3, etc.). How many new features will you create?
Solution:
We're choosing 2 features from 8 where order doesn't matter (X1×X2 = X2×X1).
C(8,2) = 8!/(2! × 6!) = (8 × 7)/2 = 28 interaction features

```python
from itertools import combinations

features = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8'] 

#generate all pairwise interactions (combination of 2)

interaction  =  list(combinations(features, 2))

print(f'total  pairwise interactions: {len(interaction)}\n')

for i, inter in enumerate(interaction,1):
    print(f"{i}.  {inter}")
    
#practical application: feature engineering in ML

import  numpy as np

X = np.random.rand(100,8)  #100 samples, 8 features

# Create new features based on pairwise interactions
X_interactions = np.zeros((X.shape[0], len(interaction)))

for i,(idx1, idx2) in enumerate(interaction):
    X_interactions[:,i] = X[:,features.index(idx1)] * X[:,features.index(idx2)]
    
print(f"Original feature shape: {X.shape}")
print(f"New feature shape with interactions: {X_interactions.shape}")
print(f'Combined shape : {(X.shape[0], X.shape[1] + X_interactions.shape[1])}')
```

Problem 4: Voting Ensemble - Majority Vote
Problem:
You have 7 binary classifiers. You want to use majority voting (at least 4 must agree).
In how many ways can exactly 4 classifiers agree on class 1?
Solution:
We need to choose which 4 out of 7 vote for class 1:
C(7,4) = 7!/(4! × 3!) = (7 × 6 × 5)/(3 × 2 × 1) = 35 ways
All majority scenarios:

Exactly 4 agree: C(7,4) = 35
Exactly 5 agree: C(7,5) = 21
Exactly 6 agree: C(7,6) = 7
All 7 agree: C(7,7) = 1

Total: 35 + 21 + 7 + 1 = 64 ways

```python
from itertools import combinations
import math

def C(n, r):
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

n_classifiers = 7

print("Majority voting scenarios:\n")

for k in range(4, 8):
    ways = C(n_classifiers, k)
    print(f"{k} out of {n_classifiers} vote for class 1: {ways} ways")

total = sum(C(n_classifiers, k) for k in range(4, 8))
print(f"\nTotal majority scenarios: {total}")

# Simulate ensemble voting
import numpy as np

np.random.seed(42)
predictions = np.random.randint(0, 2, (7, 100))  # 7 classifiers, 100 samples

# Majority vote
ensemble_pred = (predictions.sum(axis=0) >= 4).astype(int)

print(f"\nExample predictions (first 10 samples):")
print(predictions[:, :10])
print(f"Ensemble vote: {ensemble_pred[:10]}")
```

Problem 5: Neural Network Architecture Search
Problem:
You're designing a neural network with 3 hidden layers. For each layer, you can choose:

Number of neurons: [32, 64, 128, 256]
Activation: [ReLU, Tanh, Sigmoid]

How many different architectures can you create?
Solution:
Using the multiplication principle:
For each of 3 layers:

4 choices for neurons
3 choices for activation
= 4 × 3 = 12 choices per layer

Total = 12 × 12 × 12 = 1,728 architectures

```python
from itertools import product

neurons = [32, 64, 128, 256]
activations = ['ReLU', 'Tanh', 'Sigmoid']

# All combinations for one layer
layer_configs = list(product(neurons, activations))

# All combinations for 3 layers
architectures = list(product(layer_configs, repeat=3))

print(f"Configurations per layer: {len(layer_configs)}")
print(f"Total architectures (3 layers): {len(architectures)}\n")

print("First 5 architectures:")
for i, arch in enumerate(architectures[:5], 1):
    print(f"{i}. Layer1: {arch[0][0]} neurons, {arch[0][1]} | "
          f"Layer2: {arch[1][0]} neurons, {arch[1][1]} | "
          f"Layer3: {arch[2][0]} neurons, {arch[2][1]}")

print(f"\n... and {len(architectures) - 5} more architectures to explore!")
```
