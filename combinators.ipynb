```python
from itertools import combinations

features = ['Age', 'Income', 'Education', 'Gender', 'Location', 
            'Job', 'Credit_Score', 'Debt', 'Savings', 'Expenses']

# Generate all combinations of 3 features
feature_combos = list(combinations(features, 3))

print(f"Total combinations: {len(feature_combos)}")
print("\nFirst 50 combinations:")
for i, combo in enumerate(feature_combos[:50], 1):
    print(f"{i}. {combo}")

```

```python
from itertools import product
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Hyperparameter options
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 10]
}

# Calculate total combinations manually
total = 1
for param, values in param_grid.items():
    total *= len(values)
    
print(f"Total hyperparameter combinations: {total}")

# Generate all combinations
all_combos = list(product(
    param_grid['n_estimators'],
    param_grid['max_depth'],
    param_grid['min_samples_split']
))

print(f"\nFirst 5 combinations:")
for i, combo in enumerate(all_combos[:5], 1):
    print(f"{i}. n_estimators={combo[0]}, max_depth={combo[1]}, min_samples_split={combo[2]}")
```

Problem 1: Data Augmentation - Image FlipsProblem:
You have 1 image and want to create variations using:

Original
Horizontal flip
Vertical flip
Rotate 90°
Rotate 180°
Rotate 270°
How many unique images can you create if you apply exactly 2 transformations? 


```python

from itertools import permutations

transformations = [
    'Original',
    'Horizontal flip',
    'Vertical flip',
    'Rotate 90°',
    'Rotate 180°',
    'Rotate 270°'
]

# Generate all permutations of 2 transformations
#P(6,2) = 6!/(6-2)! = 6!/4! = 6 × 5 = 30 combinations
transformation_combos = list(permutations(transformations, 2))

print(f"Total unique image variations with 2 transformations: {len(transformation_combos)}\n")

for i, combo in enumerate(transformation_combos, 1):
    print(f"{i}. {combo}")
    
    
```
