# Task
Develop an image classification model for the MNIST dataset using a Convolutional Neural Network (CNN). The task involves preprocessing the data by normalizing pixel values, building and training the CNN model, visualizing sample predictions with true and predicted labels to evaluate performance, and finally, saving the trained model.

## Visualize Predictions

### Subtask:
Create a visualization to display several test images along with their true labels and the model's predicted labels, to visually inspect the model's performance.


**Reasoning**:
To visually inspect the model's performance, I will generate a grid of images from the test set, displaying each image along with its true label and the label predicted by the model.



```python
plt.figure(figsize=(15, 6))
num_images_to_display = 15

for i in range(num_images_to_display):
    plt.subplot(3, 5, i + 1) # 3 rows, 5 columns
    plt.imshow(x_test[i], cmap='gray')
    true_label = y_test[i]
    predicted_label = np.argmax(prediction[i])
    plt.title(f"True: {true_label}, Predicted: {predicted_label}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Preprocess Data

### Subtask:
Normalize the pixel values of the images to a 0-1 range and reshape the data for CNN input.


**Reasoning**:
I will convert the pixel values of `x_train` and `x_test` to float32, normalize them to a 0-1 range, and then reshape them to include a channel dimension for CNN input.



```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

print("x_train shape after preprocessing:", x_train.shape)
print("x_test shape after preprocessing:", x_test.shape)
```

## Build a Convolutional Neural Network (CNN)

### Subtask:
Develop a more advanced Convolutional Neural Network model, which is typically better suited for image classification tasks like MNIST.


**Reasoning**:
I will define a Convolutional Neural Network (CNN) model by sequentially adding convolutional, pooling, flattening, and dense layers as specified in the instructions, and then display its architecture using `model.summary()`.



```python
model_cnn = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_cnn.summary()
```

**Reasoning**:
To resolve the Keras `UserWarning` regarding `input_shape` in `Sequential` models, I will explicitly add an `Input` layer as the first element in the model definition. This aligns with Keras best practices for defining the input shape.



```python
model_cnn = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_cnn.summary()
```

**Reasoning**:
I will compile the `model_cnn` with the 'adam' optimizer, 'sparse_categorical_crossentropy' loss function, and 'accuracy' as a metric, which are standard choices for multi-class classification tasks like MNIST.

```python
model_cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

**Reasoning**:
I will train the compiled `model_cnn` using the preprocessed training data (`x_train`, `y_train`) for 10 epochs. This process will adjust the model's weights to minimize the loss and improve accuracy.

```python
history = model_cnn.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

**Reasoning**:
Now that the model is trained, I will evaluate its performance on the test dataset (`x_test`, `y_test`) to calculate the final test loss and accuracy.

```python
test_loss_cnn, test_accuracy_cnn = model_cnn.evaluate(x_test, y_test)
print(f"CNN Test Loss: {test_loss_cnn:.4f}")
print(f"CNN Test Accuracy: {test_accuracy_cnn:.4f}")
```

```python
plt.figure(figsize=(12, 5))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Plot training & validation loss values
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
